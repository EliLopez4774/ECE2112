```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
brd  = pd.read_excel("board2.xlsx")
brd
```

```python
Instru = brd.loc[(brd['Electronics']>70)&(brd['Track']=='Instrumentation')&(brd['Hometown']=='Luzon'), ['Name','GEAS','Electronics']]
Instru
```

```python
brd['Average'] = brd[['Electronics', 'GEAS', 'Math', 'Communication']].mean(axis=1)
Mindy = brd[(brd['Hometown'] == 'Mindanao')&(brd['Gender'] == 'Female')&(brd['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

```python
plt.figure(figsize=(3, 4))
plt.bar(brd['Track'], brd['Average'])
plt.title('Average Grade by Track')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(6, 4))
plt.bar(brd['Gender'], brd['Average'])
plt.title('Average Grade by Gender')
```

```python
plt.figure(figsize=(3, 4))
plt.bar(brd['Hometown'], brd['Average'])
plt.title('Average Grade by Hometown')
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(x='Track', y='Average', hue='Gender', data=brd)
plt.title('Average Grade by Track and Gender')
plt.xticks(rotation=90)
plt.legend(title='Gender')

plt.figure(figsize=(4, 4))
sns.boxplot(x='Hometown', y='Average', hue='Gender', data=brd)
plt.title('Average Grade by Hometown and Gender')
plt.legend(title='Gender')

plt.figure(figsize=(6, 6))
sns.boxplot(x='Hometown', y='Average', hue='Track', data=brd)
plt.title('Average Grade by Hometown and Track')
plt.legend(title='Tracks')
```
